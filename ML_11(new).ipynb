```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Result-Experiment-2.xlsx')
df
```

```python
grade_mapping = {
    'O': 10, 'A+': 9, 'A': 8, 'B+': 7, 'B': 6,
    'C': 5, 'D': 4, 'F': 3,'AB':0,'P':4
}
df_encoded = df.copy()
```

```python
subject_columns = [col for col in df.columns if not col.startswith('Unnamed')]

# Encode grades in each subject column
for col in subject_columns:
    df_encoded[col] = df_encoded[col].map(grade_mapping).fillna(df_encoded[col])

# Save the updated file
encoded_file_path = "/content/drive/MyDrive/Colab Notebooks/Encoded_Grades.xlsx"
df_encoded.to_excel(encoded_file_path, index=False)
print(f"Encoded grades saved to: {encoded_file_path}")
```

```python
dp = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Encoded_Grades.xlsx")
dp
```

```python
for col in subject_columns:
    dp[col] = pd.to_numeric(dp[col], errors='coerce')

# Calculate row-wise mean (horizontal mean), skipping NaN values
dp['Mean_Grade'] = dp[subject_columns].mean(axis=1, skipna=True)


# Assign class based on mean grade
def classify(mean):
    if 9 <= mean <= 10:
        return 1
    elif 6 <= mean < 9:
        return 2
    elif 4 <= mean < 6:
        return 3
    else:
        return 4

dp['Class'] = dp['Mean_Grade'].apply(classify)

# Save the updated file with mean and class
final_file_path = "/content/Encoded_Grades.xlsx"
df.to_excel(final_file_path, index=False)

print(f"Final file with mean and class saved at: {final_file_path}")
dp
```

```python
class_counts = dp['Class'].value_counts().sort_index()

# Print the counts
print("Class Distribution:")
for class_label, count in class_counts.items():
    print(f"Class {class_label}: {count} students")
```
